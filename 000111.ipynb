```python
import os
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset, Subset, random_split
from torchvision import transforms
from PIL import Image
import numpy as np
import skfuzzy as fuzz

# Custom Dataset Class
class DeepfakeDataset(Dataset):
    def __init__(self, root_dir, transform=None):
        self.root_dir = root_dir
        self.transform = transform
        self.file_paths = []
        self.labels = []
        for label, class_dir in enumerate(['fake', 'real']):
            class_dir_path = os.path.join(root_dir, class_dir)
            for file_name in os.listdir(class_dir_path):
                if file_name.endswith('.jpg') or file_name.endswith('.png'):
                    self.file_paths.append(os.path.join(class_dir_path, file_name))
                    self.labels.append(label)

    def __len__(self):
        return len(self.file_paths)

    def __getitem__(self, idx):
        img_path = self.file_paths[idx]
        image = Image.open(img_path).convert('RGB')
        label = self.labels[idx]
        if self.transform:
            image = self.transform(image)
        return image, label

# Data transformations
transform = transforms.Compose([
    transforms.Resize((64, 64)),  # Reduced image size to 64x64 to save memory
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

# Adjust the root directory to your dataset's path
train_val_dataset = DeepfakeDataset(root_dir='train', transform=transform)

# Define the CNN-LSTM model
class CNNLSTM(nn.Module):
    def __init__(self):
        super(CNNLSTM, self).__init__()
        self.conv1 = nn.Conv2d(3, 16, kernel_size=3, stride=1, padding=1)
        self.conv2 = nn.Conv2d(16, 32, kernel_size=3, stride=1, padding=1)
        self.conv3 = nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=1)
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2, padding=0)
        self.lstm = nn.LSTM(64 * 8 * 8, 128, batch_first=True)
        self.fc = nn.Linear(128, 2)
        self.dropout = nn.Dropout(0.5)

    def forward(self, x):
        x = self.pool(torch.relu(self.conv1(x)))
        x = self.pool(torch.relu(self.conv2(x)))
        x = self.pool(torch.relu(self.conv3(x)))
        x = x.view(-1, 1, 64 * 8 * 8)
        x, _ = self.lstm(x)
        x = x[:, -1, :]
        x = torch.relu(x)
        x = self.dropout(x)
        x = self.fc(x)
        return x

# Training function
def train_model(model, criterion, optimizer, dataloader, device):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0

    for images, labels in dataloader:
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * images.size(0)
        _, preds = torch.max(outputs, 1)
        correct += (preds == labels).sum().item()
        total += labels.size(0)

    epoch_loss = running_loss / total
    epoch_acc = correct / total
    return epoch_loss, epoch_acc

# Validation function
def validate_model(model, criterion, dataloader, device):
    model.eval()
    running_loss = 0.0
    correct = 0
    total = 0

    with torch.no_grad():
        for images, labels in dataloader:
            images, labels = images.to(device), labels.to(device)

            outputs = model(images)
            loss = criterion(outputs, labels)

            running_loss += loss.item() * images.size(0)
            _, preds = torch.max(outputs, 1)
            correct += (preds == labels).sum().item()
            total += labels.size(0)

    epoch_loss = running_loss / total
    epoch_acc = correct / total
    return epoch_loss, epoch_acc

# Cross-validation
def cross_validate_model(model_class, dataset, device, num_epochs=10, num_folds=10):
    fold_size = len(dataset) // num_folds
    indices = list(range(len(dataset)))
    np.random.shuffle(indices)

    fold_accuracies = []

    for fold in range(num_folds):
        val_indices = indices[fold * fold_size:(fold + 1) * fold_size]
        train_indices = list(set(indices) - set(val_indices))

        train_subset = Subset(dataset, train_indices)
        val_subset = Subset(dataset, val_indices)

        train_loader = DataLoader(train_subset, batch_size=16, shuffle=True)
        val_loader = DataLoader(val_subset, batch_size=16, shuffle=False)

        model = model_class().to(device)
        criterion = nn.CrossEntropyLoss()
        optimizer = optim.Adam(model.parameters(), lr=0.001)

        best_val_acc = 0.0

        for epoch in range(num_epochs):
            train_loss, train_acc = train_model(model, criterion, optimizer, train_loader, device)
            val_loss, val_acc = validate_model(model, criterion, val_loader, device)

            if val_acc > best_val_acc:
                best_val_acc = val_acc
                torch.save(model.state_dict(), f'best_model_fold{fold}.pth')

            print(f'Fold {fold+1}/{num_folds}, Epoch {epoch+1}/{num_epochs}')
            print(f'Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.4f}')
            print(f'Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.4f}')

        fold_accuracies.append(best_val_acc)

    return fold_accuracies

# Model training and validation
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
num_epochs = 10

# Perform cross-validation
fold_accuracies = cross_validate_model(CNNLSTM, train_val_dataset, device)
print(f'Cross-Validation Accuracies: {fold_accuracies}')
print(f'Mean Cross-Validation Accuracy: {np.mean(fold_accuracies):.4f}')
```
